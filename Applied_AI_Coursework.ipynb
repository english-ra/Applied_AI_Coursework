### Section C: Working Prototype

To begin, I need to import the required modules.

````python
from copy import deepcopy # (Copy in Python (deep copy and shallow copy) 2022)
````

Next, I am going to create the code for the Tic Tac Toe gameplay. This is what will enable the basic gameplay of the game to take place.

````python
# Used to show an easy to understand graphical format of the game board
def show_board(board, tutorial = False):
    if tutorial:
        print("     0   1   2    <- Column")
    else:
        print("     0   1   2")
    print("    - - - - - -  ")
    print(" 0 | " + board[0][0] + " | " + board[0][1] + " | " + board[0][2] + " | ")
    print("    - - - - - -  ")
    print(" 1 | " + board[1][0] + " | " + board[1][1] + " | " + board[1][2] + " | ")
    print("    - - - - - -  ")
    print(" 2 | " + board[2][0] + " | " + board[2][1] + " | " + board[2][2] + " | ")
    print("    - - - - - -  ")
    if tutorial:
        print(" ^")
        print(" |")
        print("Row")
    print()
    
    
# Used to place a game piece onto the board
def place_piece_on_board(board, col, row, piece):
    # Check that the piece is a valid game piece
    if piece != "X" and piece != "O":
        return False
    
    # Check there isn't already a piece in that position
    if board[row][col] == " ":
        # Place the piece
        board[row][col] = piece
        return True
    
    return False
    

# Used to determine who is the winner, or return None if there isn't one
def who_won(board):
    pieces = ["X", "O"]
    
    # Iterate through both game pieces
    for piece in pieces:
        # Check the rows for matching pieces
        for row in board:
            if row.count(piece) == 3:
                return piece

        # Check the columns for matching pieces
        for col in range(3):
            if board[0][col] == piece and board[1][col] == piece and board[2][col] == piece:
                return piece

        # Check the diagonals for matching pieces
        if board[0][0] == piece and board[1][1] == piece and board[2][2] == piece:
            return piece
        
        if board[0][2] == piece and board[1][1] == piece and board[2][0] == piece:
            return piece

    return None
````

Next, I created the functions to generate the inputs needed for the Minimax algorithm

````python
# This is used to evaluate the game board - the utility function
def evaluate(board):
    if who_won(board) == "X":
        return 1
    elif who_won(board) == "O":
        return -1
    else:
        return 0


# Used to calculate the amount of empty spaces left on the board
def amount_of_empty_spaces(board):
    empty_spaces = 0

    for row in board:
        for pos in row:
            if pos == " ":
                empty_spaces += 1
                
    return empty_spaces
                

# Used to calculate the available postions left on the board
def available_positions(board):
    positions = []

    # Iterate through the board and append the coordinates to the list of empty spaces
    for row_index in range(len(board)):
        for col_index in range(len(board)):
            if board[row_index][col_index] == " ":
                positions.append([col_index, row_index])

    return positions
````

Next, I created the recursive function that contains the main logic for the Minimax algorithm.

````python
def calculate_optimal_position(current_board, current_player):
    
    # X is max & O is min
    isMax = current_player == 'X'
    
    # Check whether the game is over, so that it can be evaluated and the backtracking can begin
    if amount_of_empty_spaces(current_board) == 0 or who_won(current_board) != None:
        return [evaluate(current_board), None]
    
    current_best_payoff = -float('inf') if isMax else float('inf')                                   # https://stackoverflow.com/questions/53286597/if-statement-in-python-shorthand & https://www.geeksforgeeks.org/python-infinity/
    current_best_position = None
    
    # Iterate through all potential positions where a piece could be placed
    for position in available_positions(current_board):
            
        # Create a copy of the game board, to apply potential moves to
        potential_board = deepcopy(current_board)
            
        # Place the potential piece on the board
        place_piece_on_board(potential_board, position[0], position[1], current_player)
            
        # Search for opponents best move
        opponent_piece = 'O' if isMax else 'X'
        opp_potential_best_payoff = calculate_optimal_position(potential_board, opponent_piece)[0]
            
        # If a better move is found, update it
        if isMax:
            if opp_potential_best_payoff > current_best_payoff:
                current_best_payoff = opp_potential_best_payoff
                current_best_position = position
        else:
            if opp_potential_best_payoff < current_best_payoff:
                current_best_payoff = opp_potential_best_payoff
                current_best_position = position
            
    return [current_best_payoff, current_best_position]
````

Next, I created the opponent AI agent

````python
class AI:
    def __init__(self, board, piece):
        self.board = board
        self.piece = piece
        
    def calculate_next_move(self):
        print("Player " + self.piece + ": It's my turn! Calculating...")
        next_move = calculate_optimal_position(self.board, self.piece)[1]
        print("I have decided the next move I'm going to make is Column: " + str(next_move[0]) + " & Row: " + str(next_move[1]))
        return next_move
````
